# Handling Imbalance

```python
X = df.drop(['Class'], axis=1)
y = df['Class']
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_smote, y_smote = sm.fit_resample(X, y)
```

```python
print (X.shape)
print (X_smote.shape)
print(y_smote.shape)
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, train_size=0.8, test_size=0.2, random_state=46)
```

```python
print(X_train.shape, X_test.shape)
```

# Model Training
#  -Logistic Regression

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
folds = KFold(n_splits=5, shuffle=True, random_state=4)
params = {"C": [0.01, 0.1, 1, 10, 100, 1000]}

model_logis_cv = GridSearchCV(estimator = LogisticRegression(),
                        param_grid = params, 
                        scoring= 'roc_auc', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True) 

model_logis_cv.fit(X_train, y_train)
```

```python
model_logis_results = pd.DataFrame(model_logis_cv.cv_results_)
model_logis_results
```

```python
plt.figure(figsize=(8, 6))
plt.plot(model_logis_results['param_C'], model_logis_results['mean_test_score'])
plt.plot(model_logis_results['param_C'], model_logis_results['mean_train_score'])
plt.xlabel('C')
plt.ylabel('roc_auc')
plt.legend(['test result', 'train result'], loc='upper left')
plt.xscale('log')
```

```python
model_logis_cv.best_params_
```

```python
best_score = model_logis_cv.best_score_
#best_C = model_cv.best_params_['C']

print(" The highest test is {0}".format(best_score))
```

```python
#Prediction on Test data
y_train_pred = model_logis_cv.predict(X_train)
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score, accuracy_score,confusion_matrix
```

```python
# Confusion matrix
confusion = confusion_matrix(y_train, y_train_pred)
print(confusion,end="\n\n")

# Accuracy
print("Accuracy  :",accuracy_score(y_train, y_train_pred))

#Precision
print("Precision :",precision_score(y_train, y_train_pred))

#Recall
print("Recall    :",recall_score(y_train, y_train_pred))

# F1 score
print("F1-Score  :",f1_score(y_train, y_train_pred))
```

# Prediction on Test data 

```python
y_test_pred = model_logis_cv.predict(X_test)
```

```python
# Confusion matrix
confusion = confusion_matrix(y_train, y_train_pred)
print(confusion,end="\n\n")

# Accuracy
print("Accuracy  :",accuracy_score(y_test, y_test_pred))

#Precision
print("Precision :",precision_score(y_test, y_test_pred))

#Recall
print("Recall    :",recall_score(y_test, y_test_pred))

# F1 score
print("F1-Score  :",f1_score(y_test, y_test_pred))
```

# -XGBoost

```python
# hyperparameter tuning with XGBoost

folds = 3
params = {'learning_rate': [0.2, 0.6], 
             'subsample': [0.3, 0.6, 0.9],
              'max_depth':[2,3],
               'n_estimators':[100,200,300],
            'objective':['binary:logistic']}   


model_xgb_cv = GridSearchCV(estimator = XGBClassifier(), 
                        param_grid = params, 
                        scoring= 'roc_auc', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      

# fit the model
model_xgb_cv.fit(X_train, y_train)    
```

```python
model_xgb_cv.best_params_
```

```python
#Prediction on Test data
y_train_pred = model_xgb_cv.predict(X_train)
```

```python
# Confusion matrix
confusion = confusion_matrix(y_train, y_train_pred)
print(confusion,end="\n\n")

# Accuracy
print("Accuracy  :",accuracy_score(y_train, y_train_pred))

#Precision
print("Precision :",precision_score(y_train, y_train_pred))

#Recall
print("Recall    :",recall_score(y_train, y_train_pred))

# F1 score
print("F1-Score  :",f1_score(y_train, y_train_pred))
```

# Prediction on Test data

```python
y_test_pred = model_xgb_cv.predict(X_test)
```

```python
# Confusion matrix
confusion = confusion_matrix(y_train, y_train_pred)
print(confusion,end="\n\n")

# Accuracy
print("Accuracy  :",accuracy_score(y_test, y_test_pred))

#Precision
print("Precision :",precision_score(y_test, y_test_pred))

#Recall
print("Recall    :",recall_score(y_test, y_test_pred))

# F1 score
print("F1-Score  :",f1_score(y_test, y_test_pred))
```
